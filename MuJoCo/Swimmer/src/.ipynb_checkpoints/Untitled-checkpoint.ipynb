```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_points(y):
    x= np.arange(len(y))
    
    plt.figure(figsize=(5,5))
    plt.plot(x,y)
    plt.show()
```

```python
class SineSampler():
    def __init__(self, steps_till_reset):
        self.i = 0
        self.target = 1
        self.steps = steps_till_reset
        self.b = np.arcsin(0)
        self.a = 1

    def sample(self):
        y = self.a * np.sin(self.i/self.steps * np.pi/2) + self.b
        self.i += 1

        if self.i == self.steps:
            self.b = self.target
            self.target = np.random.uniform(-1, 1, 1)
            self.a = (self.target - self.b)
            self.i = 0

        return y
```

```python
sampler = SineSampler(100)
l = [sampler.sample() for _ in range(1_000)]

draw_points(l)
```
